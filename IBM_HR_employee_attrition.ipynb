## Data analysis

### Subtask:
Calculate descriptive statistics for numerical features and identify the distribution of categorical features. Determine the percentage of attrition in the dataset.


**Reasoning**:
Calculate descriptive statistics for numerical features, analyze the distribution of categorical features, and compute the attrition percentage.



```python
# Descriptive statistics for numerical features
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns
numerical_stats = df[numerical_features].describe()
display(numerical_stats)

# Distribution of categorical features
categorical_features = df.select_dtypes(include=['object']).columns
for col in categorical_features:
    print(f"\nColumn: {col}")
    display(df[col].value_counts(normalize=True) * 100)  # Display as percentages

# Attrition percentage
attrition_percentage = (df['Attrition'].value_counts(normalize=True) * 100)['Yes']
print(f"\nAttrition Percentage: {attrition_percentage:.2f}%")
```

## Data visualization

### Subtask:
Visualize the data distributions and relationships between key features and attrition.


**Reasoning**:
Create visualizations to show the data distributions and relationships between key features and attrition, as per the instructions. This will involve histograms, bar charts, box plots, and potentially violin plots.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histograms for numerical features
numerical_features = ['Age', 'MonthlyIncome', 'YearsAtCompany', 'YearsInCurrentRole', 'TotalWorkingYears']
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_features):
    plt.subplot(2, 3, i + 1)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()

# Bar charts for categorical features
categorical_features = ['Department', 'JobRole', 'MaritalStatus', 'BusinessTravel', 'EducationField']
plt.figure(figsize=(20, 15))
for i, col in enumerate(categorical_features):
    plt.subplot(3, 2, i + 1)
    sns.countplot(x=col, hue='Attrition', data=df)
    plt.title(f'Attrition by {col}')
    plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Relationship between attrition and key features
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.boxplot(x='Attrition', y='Age', data=df)
plt.title('Age vs. Attrition')

plt.subplot(1, 3, 2)
sns.countplot(x='Department', hue='Attrition', data=df)
plt.title('Department vs. Attrition')
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 3, 3)
sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)
plt.title('Years at Company vs. Attrition')

plt.tight_layout()
plt.show()
```
